```python
file = open('./tabs2/Abusey Junction.txt').read()
file
```

```python
import re
r = re.compile('\|.+\|')
tabs = r.findall(file)
tabs
```

```python
reformated = {}
for index, tab in enumerate(tabs):
    if index%6 not in reformated:
        reformated[index%6] = []
    reformated[index%6].append(tab)
```

```python
joined = [''.join(x).replace('||', '|').replace('(', '').replace(')', '').replace('/', '') for x in reformated.values()]
joined
```

```python
import re
string = re.sub(r'([\d-]){1}h([\d-]){1}',r'\g<1>@\g<2>', file)
string = re.sub(r'([\d-]){1}p([\d-]){1}',r'\g<1>%\g<2>', file)
string = re.sub(r'[a-zA-Z]', '', string)
string = re.sub(r'@', 'h', string)
string = re.sub(r'%', 'p', string)
string = re.sub(r'[\[\]\/\\\*\#\(\)]', '', string)
string 
```
